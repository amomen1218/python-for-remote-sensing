```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
bd_shp = gpd.read_file('/content/bangladesh_division')
excel_bd = pd.read_excel('/content/bd_excel_pop/population_of_BD.xlsx')
merge_data = bd_shp.merge(excel_bd, on='ADM1_EN', how='left')
merge_data.to_file('newfile_bd.shp')
merge_data.plot(column='TOTAL POP',
                figsize=(20,10), cmap='Greys', edgecolor='black', linewidth=0.4,
              legend=True)

plt.style.use('bmh')
plt.title('Divisional Population in BANGLADESH', color='grey',fontdict={'fontsize': '20', 'fontweight': '3'})
plt.show()

```
