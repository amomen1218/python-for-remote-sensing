```python
import geopandas as gpd
import matplotlib.pyplot as plt
plt.style.use('bmh') # better for plotting geometries vs general plots.

world = gpd.read_file('/content/world_countries')
world = world.iloc[ 0:247]#there is 247(row) countris in this file
world.plot(figsize=(10,5),color='black', edgecolor='white', linewidth= 0.4)
plt.show()
```

```python
import geopandas as pd
import matplotlib.pyplot as plt
world = pd.read_file('/content/world_countries')
country_name = 'Bangladesh' #write a country name to get their boundary
country = world[world['SOVEREIGNT'] == country_name]
country.plot (figsize=(10,5), color='#196f3d', edgecolor='black', linewidth=1.0)
plt.title(f'Map of {country_name}')
plt.show()
```
